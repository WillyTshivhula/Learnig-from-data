- m = (y=data["AGE"]-y[2])/(x[1]-x[2]) #Formula to calculate slope

```python
# Regression slope & Regression intercept
#variables
X = data['CHRONIC DISEASE']
y = data['AGE']
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
regression_slope = model.params['CHRONIC DISEASE']
regression_intercept = model.params['const']


plt.figure(figsize = [10,5])
plt.title('Regression Line Distribution of Confirmed Lung Cancer vs Chronic Disease')

sns.regplot(x = data['LUNG_CANCER'], y=data['CHRONIC DISEASE'])

plt.xlabel('Distribution of lung cancer cases')
plt.ylabel("Chronic disease detect")





print(f"My Regression slope is : {regression_slope}")
print(f"My Regression intercept is : {regression_intercept}")

```

### comment 

Now let Takle hypothesis testing.

Interpreting the results. we can print the p-value and compare it to the significance level (alpha) to determine whether to reject or fail to reject the null hypothesis.


```python
# Null hypothesis: There is no difference in the mean weight between two groups.
# Alternative hypothesis: The mean weight of group A is significantly different from the mean weight of group B.

H0 = "The mean weight of group A is equal to the mean weight of group B."
Ha = "The mean weight of group A is significantly different from the mean weight of group B."



# Independent samples t-test
t_stat, p_val = stats.ttest_ind(data['LUNG_CANCER'], data['AGE'])

alpha = 0.05
if p_val < alpha:
    print("______________________________________________________________________________________")
    print(f"p-value: {p_val}. \nReject the null hypothesis.\n{Ha}")
    print("______________________________________________________________________________________")
else:
    print("______________________________________________________________________________________")
    print(f"p-value: {p_val}. Fail to reject the null hypothesis. \n{H0}")
    print("______________________________________________________________________________________")
```

### Comment 
Now lets do the Chi-squared test

```python

from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(data['AGE'], data['LUNG_CANCER'])
stat, p, dof, expected = chi2_contingency(contingency_table)
print('Chi-Squared Statistic: ', stat)
print('Degrees of Freedom: ', dof)
print('P-value: ', p)




```

```python
from scipy.stats import ttest_ind
group1 = data['AGE'].values
group2 = data['LUNG_CANCER'].values

stat, p = ttest_ind(group1, group2)

print('T-Statistic: ', stat)
print('P-value: ', p)

```
